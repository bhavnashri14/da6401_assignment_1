**2.1 Data Exploration and Class Distribution**

```python
import wandb
import numpy as np
import matplotlib.pyplot as plt
from utils.data_loader import load_data


# Initialize W&B
wandb.init(project="da6401_assignment1", name="mnist_data_exploration")

# Load MNIST
X_train, y_train, _, _ = load_data("mnist")

# Convert one-hot to labels
labels = np.argmax(y_train, axis=1)

# Create W&B table
table = wandb.Table(columns=["Class", "Image"])

# For each digit 0–9
for digit in range(10):
    idx = np.where(labels == digit)[0][:5]  # first 5 samples

    for i in idx:
        img = X_train[i].reshape(28, 28)
        table.add_data(
            digit,
            wandb.Image(img, caption=f"Digit: {digit}")
        )

# Log table
wandb.log({"MNIST Sample Images": table})

wandb.finish()
```
